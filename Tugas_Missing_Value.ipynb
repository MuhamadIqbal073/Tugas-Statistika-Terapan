```python
#Menampilkan missing values, yang dimana pada reading score terdapat 1 missing values
print(dataset.isnull().sum())
```

```python
#Menghilangkan missing values menggunakan drop, yang dimana metode drop menghilangkan baris yang terdapat NaN
#Dapat dilihat pada tabel baris 2 yang terdapat NaN menghilang dikarenakan sudah di drop
dataset.dropna(inplace=True)
print(dataset.isnull().sum())

dataset.head()
```

```python
#HANDLING MISSING VALUES USING MEAN
#Mengganti nilai NaN pada reading score menggunakan mean
dataset["reading score"] = dataset["reading score"].replace(num.NaN, dataset["reading score"].mean())
print(dataset["reading score"][:5]) 

dataset.head()
```

```python
#Nilai yang mengisi NaN pada kolom reading score dengan menggunakan mean adalah 68.4274193548387
mean = dataset["reading score"].mean()
mean
```

```python
#HANDLING MISSING VALUES USING MEDIAN
#Mengganti nilai NaN pada reading score menggunakan median
dataset["reading score"] = dataset["reading score"].replace(num.NaN, dataset["reading score"].median())
print(dataset["reading score"][:5])
dataset.head()
```

```python
#Nilai yang mengisi NaN pada kolom reading score dengan menggunakan median adalah 69.0
median = dataset["reading score"].median()
median
```

```python
#HANDLING MISSING VALUES USING MODE
#Mengganti nilai NaN pada reading score menggunakan modus
dataset["reading score"] = dataset["reading score"].replace(num.NaN, statistics.mode(dataset["reading score"]))
print(dataset["reading score"][:5])

dataset.head()
```

```python
#Nilai yang mengisi NaN pada kolom reading score dengan menggunakan modus adalah 63.0
modus = dataset["reading score"].mode()
modus
```

```python
#Menampilkan dataset yang telah bebas dari missing values, dapat dilihat missing values pada reading score telah menghilang 
print(dataset.isnull().sum())
```
